## Regroupement des tailles de pizzas

```python
dataset['Size']
```

## Analyse du prix des pizzas en fonction de leur taille

```python
regroupement_prix_taille = dataset.groupby('Size').agg({'Price': ['mean', 'min', 'max']})

print(regroupement_prix_taille)
```

```python
ax = regroupement_prix_taille.plot(kind='bar', figsize=(15,7))
ax.set(xlabel='Size', ylabel='Price')
ax.legend(title='Price values')
```

## Graphique montrant le prix moyen d'une pizza en fonction de sa taille

```python
plt.figure(figsize=(20,7))
sns.barplot(dataset['Size'], dataset['Price'])
```

Nous observons ici que comme le veut la logique la pizza la plus petite est la moins chère alors que la plus grande est la plus onéreuse

## Graphique représentant le prix moyen d'une pizza en fonction de sa taille pour chaque restaurant 

```python
ax = dataset.pivot_table(values='Price', index='Company', columns='Size').plot(kind='bar', figsize=(25,7))
ax.set(xlabel='Companies', ylabel='Price')
ax.legend(title='Size')
```

## Nuage de mots mettant en avant les mots les plus utilisés dans la description des pizzas

```python
from wordcloud import WordCloud, STOPWORDS

df1 = dataset.rename(columns={'Pizza Name':'Pizza_Name'})
comment_words = ''
stopwords = set(STOPWORDS)

for val in df1.Pizza_Name:
    val = str(val)
    tokens = val.split()
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
        
    comment_words += " ".join(tokens)+" "
    
wordcloud = WordCloud(width = 1200, height = 1200, 
                     background_color='white',
                     stopwords = stopwords,
                     min_font_size = 10).generate(comment_words)

plt.figure(figsize = (8,8), facecolor = None)
plt.imshow(wordcloud)
plt.axis('off')
plt.tight_layout(pad = 0)

plt.show()
```
